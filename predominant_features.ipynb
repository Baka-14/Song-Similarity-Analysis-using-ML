```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
import math

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('output_data.xlsx')
```

```python
#df.head()
df.head()
```

```python
df.dropna(inplace=True)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# Compute information gain for each feature
ig = []
for feature in df.columns:
    entropy_before = -sum([p*np.log2(p) for p in df.iloc[:,-1].value_counts(normalize=True)])
    entropy_after = 0
    for value in df[feature].unique():
        p = len(df[df[feature]==value])/len(df)
        if p != 0:
            entropy_after += -p*math.log2(p)
    ig.append((feature, entropy_before - entropy_after))

# Compute Pearson correlation coefficient for each feature
pearson_corr = []
for feature in df.columns:
    corr, _ = pearsonr(df[feature], df.iloc[:,0])
    pearson_corr.append((feature, corr))

# Compute Spearman correlation coefficient for each feature
spearman_corr = []
for feature in df.columns:
    corr, _ = spearmanr(df[feature], df.iloc[:,0])
    spearman_corr.append((feature, corr))

# Sort the results by the magnitude of the computed coefficients
ig_sorted = sorted(ig, key=lambda x: x[1], reverse=True)
pearson_corr_sorted = sorted(pearson_corr, key=lambda x: abs(x[1]), reverse=True)
spearman_corr_sorted = sorted(spearman_corr, key=lambda x: abs(x[1]), reverse=True)

# Print the results
print("Information Gain:")
for f, ig_val in ig_sorted:
    print(f"{f}: {ig_val:.3f}")

```

```python
print("\nPearson Correlation Coefficient:")
for f, corr_val in pearson_corr_sorted:
    print(f"{f}: {corr_val:.3f}")
```

```python

print("\nSpearman Correlation Coefficient:")
for f, corr_val in spearman_corr_sorted:
    print(f"{f}: {corr_val:.3f}")
```

```python
with open("spearman_correlation.txt", "r") as f:
    # Create an empty list to store the feature names
    features = []
    
    for line in f:
        # Check if the line contains a colon and split the line
        if ":" in line:
            line_split = line.split(":")
            feature_name = line_split[0].strip()  # Get the feature name
            value = line_split[-1].strip()  # Get the feature value
            
            # Check if the value is -1 and store the feature name in the list
            if value == '-1.000':
                features.append(feature_name)

# Print the list of feature names with value -1
print(features)
```

```python

with open('features.txt', 'w') as f:
    for item in features:
        f.write("%s\n" % item)
```

```python
import pandas as pd
from sklearn.manifold import TSNE
import seaborn as sns
import numpy as np
from fcmeans import FCM

# Load data
df = pd.read_excel('rock.xlsx')
#df.dropna(inplace=True)
#df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.head()
```

```python

# Normalize the data
data_norm = (df - df.mean()) / df.std()

# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
tsne_results = tsne.fit_transform(data_norm)

# Perform Fuzzy C-Means clustering
fcm = FCM(n_clusters=5)
fcm.fit(data_norm)
fcm_centers = fcm.centers
fcm_labels = fcm.predict(data_norm)

# Visualize clusters
palette = sns.color_palette('deep', 5)
cluster_colors = [palette[i] for i in fcm_labels]
sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], hue=fcm_labels, legend='full', palette=cluster_colors)
```

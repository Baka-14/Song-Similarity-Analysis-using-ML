#### DB SCAN Clustering

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Configuration options
num_samples_total = 1000
cluster_centers = [(3,3), (7,7)]
num_classes = len(cluster_centers)
epsilon = 1.0
min_samples = 13

# Generate data
X, y = make_blobs(n_samples = num_samples_total, centers = cluster_centers, n_features = num_classes, center_box=(0, 1), cluster_std = 0.5)

np.save('./clusters.npy', X)
X = np.load('./clusters.npy')

# Compute DBSCAN
db = DBSCAN(eps=epsilon, min_samples=min_samples).fit(X)
labels = db.labels_

no_clusters = len(np.unique(labels) )
no_noise = np.sum(np.array(labels) == -1, axis=0)

print('Estimated no. of clusters: %d' % no_clusters)
print('Estimated no. of noise points: %d' % no_noise)

# Generate scatter plot for training data
colors = list(map(lambda x: '#3b4cc0' if x == 1 else '#b40426', labels))
plt.scatter(X[:,0], X[:,1], c=colors, marker="o", picker=True)
plt.title('Two clusters with data')
plt.xlabel('Axis X[0]')
plt.ylabel('Axis X[1]')
plt.show()
```

### Fuzzy C Means clustering 

```python
import numpy as np
import matplotlib.pyplot as plt
from fcmeans import FCM
from matplotlib.cm import get_cmap

# Generate sample dataset
np.random.seed(42)
X = np.random.rand(2000, 2) * 10  # Modified sample dataset with larger values

# Build the fuzzy C-means clustering model
fcm = FCM(n_clusters=10)  # Specify the number of clusters
fcm.fit(X)

# Get the predicted cluster labels
labels = fcm.predict(X)

# Get the cluster centers
centers = fcm.centers

# Plot the data points with cluster colors
cmap = get_cmap('tab10')
colors = cmap(labels.astype(float) / fcm.n_clusters)

plt.scatter(X[:, 0], X[:, 1], marker='o', color=colors, edgecolors='k')

# Plot the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='x', color='red', linewidth=2)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Fuzzy C-means Clustering')

plt.show()
```
